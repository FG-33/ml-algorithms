```python
# define functions for training data generation
import numpy as np

# linear
def f(x):
    return 23.5 * x + 133

# linear
def g_with_noise(x):
    return -15.4 * x - 117 + np.random.random_sample() * 10

# quadratic
def h(x):
    return x * x + 10

# polinomial
def i(x):
    return 4.633858315*10**(-10)*x**7 - 1.901678734*10**(-7)*x**6 + 3.019149671*10**(-5)*x**5 - 2.351223031*10**(-3)*x**4 + 9.355836155*10**(-2)*x**3 - 1.791650563*x**2 + 14.35996841*x - 10.65958431
```

```python
# generate two random lists of input values for the functions above
np.random.seed(0)

x_f = np.random.random_sample((50,)) * 200 - 100
x_g = np.random.random_sample((50,)) * 200 - 100
x_h = np.random.random_sample((50,)) * 200 - 100
x_i = np.random.random_sample((50,)) * 110
```

```python
# plot f
import matplotlib.pyplot as plt
plt.figure(1, figsize=(15, 5))
plt.xlabel("x")
plt.ylabel("f(x_f)")
y_f = f(x_f)
plt.plot(x_f, y_f, "rx")
plt.plot(x_f, y_f, "b-")
plt.show()
```

```python
# write y_f to file
with open("f.data", 'w') as out_file:
    for ij in range(x_f.shape[0]):
        out_file.write("{},{}\n".format(y_f[ij], x_f[ij]))
```

```python
# plot g_with_noise
plt.figure(1, figsize=(15, 5))
plt.xlabel("x_g")
plt.ylabel("g_with_noise(x_g)")
y_g = g_with_noise(x_g)
plt.plot(x_g, y_g, "rx")
plt.plot(x_g, y_g, "b-")
plt.show()
```

```python
# write y_g to file
with open("g.data", 'w') as out_file:
    for ij in range(x_g.shape[0]):
        out_file.write("{},{}\n".format(y_g[ij], x_g[ij]))
```

```python
# plot h
plt.figure(1, figsize=(15, 5))
plt.xlabel("x_h")
plt.ylabel("h(x_h)")
y_h = h(x_h)
plt.plot(x_h, y_h, "rx")
x_h_sorted, y_h_sorted = zip( *sorted( zip(x_h, y_h) ) ) # do this for a pretty plot
plt.plot(x_h_sorted, y_h_sorted, "b-")
plt.show()
```

```python
# write y_h to file
with open("h.data", 'w') as out_file:
    for ij in range(x_h.shape[0]):
        out_file.write("{},{}\n".format(y_h[ij], x_h[ij]))
```

```python
# plot h
plt.figure(1, figsize=(15, 5))
plt.xlabel("x_i")
plt.ylabel("i(x_i)")
y_i = i(x_i)
plt.plot(x_i, y_i, "rx")
x_i_sorted, y_i_sorted = zip( *sorted( zip(x_i, y_i) ) ) # do this for a pretty plot
plt.plot(x_i_sorted, y_i_sorted, "b-")
plt.show()
```

```python
# write y_i to file
with open("i.data", 'w') as out_file:
    for ij in range(x_i.shape[0]):
        out_file.write("{},{}\n".format(y_i[ij], x_i[ij]))
```
